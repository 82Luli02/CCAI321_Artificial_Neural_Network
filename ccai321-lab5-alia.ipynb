**Implement a Hebb learning rule as described above. Name the function “hebb_rule”. Let the function take T and p as parameters. T refers to the target value and p is the data point. The training rule updates.**

```python
def hebb_rule(t,p):
    p=np.transpose(p)
    a=np.dot(t,p)
    return a 
```

**Implement a function that normalizes a vector if it is not of unit length. Name it “normalizevec”.**

```python
def normalizevec(a):
    d=np.sqrt(np.sum(np.square(a)))
    c=a/d
    return c
```

**Test the function “normalizevec” by passing the following vector to it.**

```python
p1 = [1,-1,-1]
n=normalizevec(p1)
print('The normalized vector: {}\nThe length after of the normalized vector = {}'.format(n,check_unitlength(n)))
```
