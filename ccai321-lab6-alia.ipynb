# Q1
**Implement the first layer and name it “logsiglayer”. The function takes the following parameters: W1, p and b1. W is the weight matrix, p is the input and b is the bias.**

```python
def logsig(x):
    '''
       This function takes the value of x as an input, 
       and returns the value of the calculated logsigmoid of that number.
    '''
    return 1 / (1 + np.exp(-x))

def logsiglayer(W1,p,b1): 
    n=(np.dot(W1,p))+b1
    return logsig(n)
```

# Q2
**Implement the second layer and name it “linearlayer”. The function takes the following parameters: W2, a1 and b2. W is the weight matrix, a is the input and b is the bias.**    

```python
def purelin(x):
    '''
       This function takes the value of x as an input, 
       and returns the linear value of that number (same value).
    '''
    return x

def linearlayer(W2,a1,b2): 
    return purelin((np.dot(W2,a1))+b2)
```

```python
W1=np.array([[10],[10]])
b1=np.array([[-10],[10]])
W2=np.array([[1,1]])
b2=np.array([[0]])
p = np.array([[2]]) 
print("p = {}".format(p))
print(p.shape)
print(W1.shape)
a1=logsiglayer(W1,p,b1)
print(a1)
print(a1.shape)
a2=linearlayer(W2,a1,b2)
print("w2")
print(W2.shape)
print(a2)
```
