```python
def logsig(x):
 '''
 This function takes the value of x as an input, 
 and returns the value of the calculated logsigmoid of that number.
 '''
 return 1 / (1 + np.exp(-x))

def logsiglayer(W1,p,b1):
    print("LOgsig")
    print(logsig((np.dot(W1,p))+b1))
    return logsig((np.dot(W1,p))+b1)
```

```python
def purelin(x):
 '''
 This function takes the value of x as an input, 
 and returns the linear value of that number (same value).
 '''
 return x

def linearlayer(W2,a1,b2): 
    print("lin")
    print(purelin((np.dot(W2,a1))+b2))
    return purelin((np.dot(W2,a1))+b2)
```

```python
def forwardpropagation(p,W1,b1,W2,b2):
    a0=p
    a1=logsiglayer(W1,p,b1)
    a2=linearlayer(W2,a1,b2)
    print("a1= {}\na2 = {}".format(a1,a2))
    return a0,a1,a2
```

```python
def backpropagation(F2,t,a,F1,W2):
    s2 = -2*F2*(t-a) 
    s1 = F1*np.transpose(W2) *s2 
    print("s1= {}\ns2 = {}".format(s1,s2))
    return s2,s1
```

```python
def updateparams(W1, W2, b1, b2, s1, s2, a0, a1, a2, alpha):
    W1 = W2 - alpha * s1 * np.transpose(a0)    
    b1 = b1 - alpha * s1 
    
    W2 = W2 - alpha * (np.dot(s2,np.transpose(a1)))     
    b2 = b2 - alpha * s2
    return W1,b1,W2,b2
```

```python
def backpropagatealgorithm(W1, W2, b1, b2, p, t, alpha):
    a0,a1,a2=forwardpropagation(p,W1,b1,W2,b2)
    F1=dlogsig(a1)
    F1=np.diag(F1)
    F2=dpurelin(a2)
    s2,s1=backpropagation(F2,t,a2,F1,W2)
    W1,b1,W2,b2=updateparams(W1, W2, b1, b2, s1, s2, a0, a1, a2, alpha)
    return W1,b1,W2,b2
```
