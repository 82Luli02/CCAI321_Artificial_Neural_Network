# Q1
**Plot the output of the logsigmoid function for n=-5 to 5**

```python
#Fill array x with values ranging from -5 to 5, which will be used throughout the questions
x = np.array(np.arange(-5, 6, 1))

#Print array x
print("array x = {}".format(x))
```

```python
def logsig(x):
    '''
       This function takes the value of x as an input, 
       and returns the value of the calculated logsigmoid of that number.
    '''
    return 1 / (1 + np.exp(-x))
```

```python
#Fill array y with the values returned using logsigmoid function
y = np.array([logsig(x[i]) for i in range(x.size)])

#Print array y
print("\narray y = {}\n".format(y))

#Plot the coordinates 
plt.plot(x,y)
```

# Q2
**Plot the output of the hardlim transfer function for n=-5 to 5**

```python
def hardlim(x):
    '''
       This function takes the value of x as an input. 
       - returns: 0 if x is less than 0, 1 otherwise.  
    '''
    if x<0:
        return 0
    return 1
```

```python
#Fill array y with the values returned using hardlimit transfer
y = np.array([hardlim(x[i]) for i in range(x.size)])

#Print array y
print("\narray y = {}\n".format(y))

#Plot the coordinates 
plt.plot(x,y)
```

# Q4
**Plot the output of the the Hyperbolic Tangent transfer function for n=-5 to 5**

```python
def htan(x):
    '''
       This function takes the value of x as an input, 
       and returns the value of the calculated hyperbolic tangent of that number.
    '''
    return(np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))
```

```python
#Fill array y with the values returned using hyperbolic tangent transfer. Also, we can use "math.tanh" instead.
y = np.array([htan(x[i]) for i in range(x.size)])

#Print array y
print("\narray y = {}\n".format(y))

#Plot the coordinates 
plt.plot(x,y)
```

# Q5
**Plot the output of the positive linear transfer function for n=-5 to 5**

```python
def poslin(x):
    '''
       This function takes the value of x as an input.
       - returns: 0 if x is less than 0, x otherwise.  
    '''
    if x<0:
        return 0
    return x
```

```python
#Fill array y with values returned using positive linear transfer
y = np.array([poslin(x[i]) for i in range(x.size)])

#Print array y
print("\narray y = {}\n".format(y))

#Plot the coordinates 
plt.plot(x,y)
```
