# Transfer functions

```python
def logsig(x):
    '''
       This function takes the value of x as an input, 
       and returns the value of the calculated logsigmoid of that number.
    '''
    return 1 / (1 + np.exp(-x))
```

```python
def hardlim(x):
    '''
       This function takes the value of x as an input. 
       - returns: 0 if x is less than 0, 1 otherwise.  
    '''
    if x<0:
        return 0
    return 1
```

```python
def purelin(x):
    '''
       This function takes the value of x as an input, 
       and returns the linear value of that number (same value).
    '''
    return x
```

```python
def htan(x):
    '''
       This function takes the value of x as an input, 
       and returns the value of the calculated hyperbolic tangent of that number.
    '''
    return(np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))
```

```python
def poslin(x):
    '''
       This function takes the value of x as an input.
       - returns: 0 if x is less than 0, x otherwise.  
    '''
    if x<0:
        return 0
    return x
```

# Q1
**Develop a single input neuron. Assuming that it has an input p, weight w, bias b, summation output n, transfer function f and output a.**

**The steps for implementation require you to first calculate `wp`.**

**Two numbers can be multiplied in Python using the following command:** `4*6` **or for variables:** `x*y`

**Next, calculate the sum of `wp` and `b` and this is the output of the summation, i.e.,** `n = wp + b`

**Finally, use this as input to the transfer function to get your result.**

**Please create a function titled `single_input_neuron` and pass the parameters i.e., values of `p`, `w`, `b`, and `f` (i.e., name of transfer function) and return the output `a` from that function.**


```python
def single_input_neuron(p,w,b,f):
    a= (p*w)+b
    if(f=='logsig'):
        return logsig(a)
    elif(f=='hardlim'):
        return hardlim(a)
    elif(f=='poslin'):
        return poslin(a)
    elif(f=='purelin'):
        return purelin(a)
    elif(f=='htan'):
        return htan(a)
```

```python
print('After applying hardlim function = {}'.format(single_input_neuron(2,-1.5,3,'hardlim')))
print('After applying logsig function = {}'.format(single_input_neuron(2,-1.5,3,'logsig')))
print('After applying poslin function = {}'.format(single_input_neuron(2,-1.5,3,'poslin')))
```

# Q2
**The steps for implementation require you to first calculate `wp`. This time both `w` and `p` are either vectors or matrices i.e.** 

**p = \begin{bmatrix} 1 \\ 2 \\ 3 \end{bmatrix}**

**w =  \begin{bmatrix} 4 & 5 & 6 \end{bmatrix}**

**Assuming that `p` and `w` are vectors. Matrix or vector multiplication can be done in Python using the NumPy library function `dot`, i.e., `np.dot(w, p)` (Note: Inputs are provided as arrays and the transpose is taken by Python itself)**

**Next, calculate the sum of `wp` and `b` and this is the output of the summation, i.e., `n = np.dot(w, p) + b`**

**Finally, use this as input to the transfer function to get your result.**

**Please create a function titled `neuron` and pass the parameters i.e., values of ‘p’, ‘w’, ‘b’, and ‘f’ i.e., name of transfer function and return the output ‘a’ from that function.**

```python
def multiple_input_neuron(p,w,b,f):
    a= np.dot(p,w)+b
    if(f=='logsig'):
        return logsig(a)
    elif(f=='hardlim'):
        return hardlim(a)
    elif(f=='poslin'):
        return poslin(a)
    elif(f=='purelin'):
        return purelin(a)
    elif(f=='htan'):
        return htan(a)
```

**What the output ‘a’ of the multiple input neuron if  p=[1 2 3], w=[4 5 6], b=-1.5 and transfer function is ‘poslin’?**

```python
p =np.array([1,2,3])
w = np.array([4,5,6])
print('Vector p = {}'.format(p))
print('Vector w = {}'.format(w))
print('After applying poslin function = {}'.format(multiple_input_neuron(p,w,-1.5,'poslin')))
```
