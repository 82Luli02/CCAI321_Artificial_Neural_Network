# Transfer function

```python
def hardlim(x):
    '''
       This function takes the value of x as an input. 
       - returns: 0 if x is less than 0, 1 otherwise.  
    '''
    if x<0:
        return 0
    return 1
```

# Q1
**Implement a single neuron perceptron model, with a hardlim transfer function.**

```python
def single_neuron_perceptron(p,w,b):
    return hardlim(np.dot(w,p)+b)
```

# Q2
**Implement a perceptron learning rule function as described above. Name the function 
“learning_rule”. Let the function take t, a, w, b and p as parameters. t refers to the target value, a refers to the predicted value, w and b are the old weight and bias respectively, and p is the data point.** 


```python
def learning_rule(t,a,w,b,p):
    '''
    Inputs:
       t -> refers to the target value,
       a -> refers to the predicted value,
       w -> old weight,
       b -> old bias,
       p -> data point
    '''
    e = t-a
    w_new = w + (np.dot(e,p))
    b_new = b + e
    return w_new,b_new
```

```python
w=np.random.rand(1,2)
b=np.random.rand(1,1)

p=[[-2,2],[-2,0],[-2,-2],[0,2],[0,-2],[2,2],[2,0],[2,-2]]
t=[1,1,1,1,0,0,0,0]

for i in range (0,8):
    print(">> w{}: {}\tb{}: {}".format(i+1,w,i+1,b))
    a=single_neuron_perceptron(p[i],w,b)
    print("- a{}: {}\n".format(i+1,a))
    w,b=learning_rule(t[i],a,w,b,p[i])
```
